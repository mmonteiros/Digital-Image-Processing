#### (1) - Create the image of a paraboloid with one axis scaled (like an oval paraboloid).

```python
import numpy as np
import matplotlib.pyplot as plt

def paraboloid(x, y, a, b, c):
    return c*((pow(x,2)/pow(a,2)) - (pow(y,2)/pow(b,2)))

X, Y = np.meshgrid(np.linspace(-7, 7, 100), np.linspace(-7, 7, 100))
Z = paraboloid(X, Y, 1, 1, 1)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 120, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
```

#### (2) Create the image of a rotated sin(x) using rotation of coordinates.

```python
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure()
ax = plt.axes(projection='3d')

Xaxis = np.linspace(-2*np.pi, 2*np.pi, 200)
Yaxis = np.linspace(-2*np.pi, 2*np.pi, 200)
tau, phase = np.meshgrid(Xaxis, Yaxis)

ax.contour3D(Xaxis, Yaxis, np.sin(tau + phase), 120, cmap='binary')
ax.set_xlabel(r'$\tau$')
ax.set_ylabel(r'$\phi$')
ax.set_zlabel('amplitude')
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

# Using the rotation of coordinates
# x' = x*cos(theta) + y*sin(theta)
# y' = -x*sin(theta) + y*cos(theta) 	
 
tau_rot = np.cos(0.50)*tau + np.sin(0.50)*phase 
phase_rot = -np.sin(0.50)*tau + np.cos(0.50)*phase

ax.contour3D(tau_rot, phase_rot, np.sin(tau + phase), 120, cmap='binary')

ax.set_xlabel(r'$\tau$')
ax.set_ylabel(r'$\phi$')
ax.set_zlabel('amplitude')
```

#### (3) Create the image of a gaussian in 2D.

```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
N = 10
Gau1d = signal.gaussian(N, std=1).reshape(N, 1)
resultKernel = np.outer(Gau1d, Gau1d)
plt.imshow(resultKernel)
plt.show()
```
